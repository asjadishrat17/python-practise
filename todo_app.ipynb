```python
#todo app 
#1 for add task
#2 for view task 
#3 for delete task
#4 exit

def add_task(task_list):
    task_index=len(task_list)+1
    task_name=input("task {}.".format(task_index))
    task_list.update({task_index : task_name})
    print("succesfully add task")

def view_task(task_list):
    for taskIndex,taskName in task_list.items():
        if taskIndex > 0:
            print("{}.{}".format(taskIndex,taskName))
        else:
            print("no data found")

def delete_task(task_list):
    del_index=int(input("which task you want to delete ?? eg:1"))
    if len(task_list)>0:
        task_list.pop(del_index)
        print("successfully delete {}".format(del_index))
    else:
        print("no data found")

def isInteger(userInput):
    try:
        int(userInput)
        return True
    except ValueError:
        return False

def menu_item(menu):
    print("======== menu ========")
    for menuIndex,menuName in menu.items():
        print("{}.{}".format(menuIndex,menuName))
    userInput=int(input("select any item\n"))
    return userInput

menu={1:"add task",
     2:"view task",
     3:"delete task",
     4:"exit"}

task_list={}

while(True):
    userInput=menu_item(menu)
    if isInteger(userInput)== True:
        userInput=int(userInput)
        if userInput > 0 and userInput <= 4:
            if userInput == 1:
                add_task(task_list)
            elif userInput==2:
                view_task(task_list)
            elif userInput==3:
                delete_task(task_list)
            elif userInput==4:
                 break;
            else:
                print("you enter wrong choice")
        else:
            print("value mus")
    else:
        print("invalid value , value must be integer")
```
